Introduction

Content 

This paper is about a face recognition technique developed by the authors by using a pattern classification approach. In this paper, a new approach to face recognition is introduced and explained briefly. Also, comparisons are made between the face recognition algorithm developed by the authors "Fisherface" and the “Eigenface” technique. After experimenting on the Yale and Harvard face databases comprehensively, it was found that the "Fisherface" technique is more accurate than the "Eigenface" technique. 
Their approach is based on the following observations: 
a.	From a fixed viewpoint, the images of a Lambertian surface under different illumination lie in a three-dimensional subspace of a high dimensional space.  
b.	The above statement is not valid if there exists region of shadow, regular reflections and various expressions that deviates from actual image resulting in less reliable for recognition. 
By utilizing these observations authors tried to find different projection of faces from high dimensional image which exclude variation in lighting and facial expressions to notably lower dimensional space. Here to achieve this the authors derived an algorithm named “Fisherface” which is based on Fisher’s Linear Discriminate (FLD) where it elucidates how to increase the ratio of between class scatter matrix to that of within class scatter matrix. 
In addition to that, the article describes Eigenface approach to achieve the abovementioned observations. This approach is based on linearly projecting image space using Principal Component Analysis (PCA) to lower dimension image space. 
 
Innovation 

The Algorithm proposed by the authors is logically explained and derived from FLD. Looking back at the time the paper was written, it was original and innovative as they took the existing Fisher’s Linear Discriminant and used it to derive their “Fisherface” algorithm. 
They have contributed a new algorithm for facial recognition. Further they compared the accuracy of Fisherface and Eigenface algorithm by performing test on face databases of Yale and Harvard. 
This article describes four appropriate approaches to address abovementioned observations and compared each method to achieve utmost accuracy in face recognition problem. These approaches include:  
a.	Correlation: It is one of the simplest classification techniques which uses nearest neighbor algorithm where it recognizes the nearest point in the labelled learning dataset and dataset refers to image space.  Best image space is said to be correlated with the test data if it has a unit variance and zero normalization. This specifies that it is insensitive to light variance and facial expressions. The limitation of the correlation is high computation time and it requires large storage to achieve facial recognition system. 
b.	Eigenface: Commonly used technique to reduce dimensionality in computer vision. This is the method which employs Principal Component Analysis (PCA) to achieve face recognition. PCA technique maximizes the scatter of all projection. Drawback of using PCA is that while increasing between class scatter it also increases within class scatter as well. This leads to increase in unwanted information which is later discarded in PCA. 
c.	Linear Subspaces: Both techniques mentioned above do not exploit the observations made and hence fail to achieve facial recognition problem. So linear subspace which utilizes the observation and considered to be good technique for a Lambertian surface without shadowing. This is a well-known method called as photo metrics stereo in which it considers image as a Lambertian surface where it takes 3 or more images of same dimensionality in high dimension as a training dataset. To compute recognition problem, it measures shortest distance to the image space. This is considered as linear subspace method. The drawback is that it only works when the given surface is a Lambertian surface.  
d.	Fisherface: It is one of the well-known methods which takes the benefits of the Eigenface and linear subspace. Fisherface technique uses the labelled training dataset to sense the information and reduces dimensionality to get more reliable method in high dimension space. This technique employs class specific method, which shapes the scatter to make it more reliable classification.    


To evaluate the class specific linear projection article supported a comparison of PCA and FLD using two class problem. In this comparison they have taken samples from each class and randomly selected in direction orthogonal to linear subspace. In this illustration both techniques projected the points from 2D to 1D. The PCA smears the classes and they are no longer linearly separable. PCA achieves a maximum total scatter whereas FLD achieves better between class scatter than within class scatter. The PCA technique is always singular and the pixel image is larger than the number of images in the learning set. Fisherface is a better computation technique to overcome this criterion. It sets actual images in lower dimensional space resulting in non-singular within class scatter.
